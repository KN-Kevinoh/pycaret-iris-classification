# Import module

```python
from pycaret import classification
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

# Load iris dataset

```python
dataset_iris = datasets.load_iris() 
```

```python
# let check classes for classification
dataset_iris.target_names
```

```python
# get data and targets
data = dataset_iris.data
labels = dataset_iris.target
```

```python
data, labels
```

```python
data.shape
```

```python
len(labels)
```

# Split data into train and test set

```python
X_train, X_test, Y_train, Y_test = train_test_split(data, labels, train_size=0.7, random_state = 42)
```

```python
X_train.shape, len(Y_train), X_test.shape, len(Y_test)
```

# Prepare our data for training and prediction, by creating dataframe structures

```python
dataset_iris.feature_names
```

```python
train_data = pd.DataFrame(X_train, columns=dataset_iris.feature_names)
test_data = pd.DataFrame(X_test, columns=dataset_iris.feature_names)
```

```python
# 'setosa': 0, 'versicolor': 1, 'virginica': 2
dataset_iris.target_names
```

```python
# labels columns
train_data['labels'] = Y_train
test_data['labels'] = Y_test
```

```python
train_data.head()
```

```python
test_data.head()
```

# Setup environment and models fitting

```python
classifier = classification.setup(train_data, target='labels')
```

## Noticed that for each k-fold the evaluation metrics will be  Accuracy, AUC, Recall, Precision, F1 and Kappa (use to compare how the predicted class closest to ground thruth)

###  Matthews correlation coefficient (MCC) or phi coefficient 

```python
# first use decision tree model
dt = classification.create_model('dt')
```

```python
# use lightgbm model ( Light Gradient Boosting - Microsoft )
lgb = classification.create_model('lightgbm')
```

### As we can see the both models have same accuracy

```python
# use ensemble models
bagging_dt = classification.ensemble_model(dt, method='Bagging')
```

```python
bagging_lgb = classification.ensemble_model(lgb, method='Bagging')
```
